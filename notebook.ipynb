## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perbandingan kualitas udara di Dongsi pada tahun 2015 dan 2016 berdasarkan kadar PM2.5?

 Visualisasi menunjukkan rata-rata kadar PM2.5 di station Dongsi pada tahun 2016 memiliki nilai yang lebih kecil dibandingkan pada tahun 2015

```python
fig, ax = plt.subplots(figsize=(15, 6))

colors = ['#D3D3D3', '#72BCD4']

sns.barplot(x='year', y='PM2.5', data=average_dongsi_pm25_df, palette=colors)
ax.set_ylabel('PM2.5')
ax.set_xlabel('Year', fontsize=15)
ax.set_title('Average PM2.5 in Dongsi', loc='center', fontsize=20)
ax.tick_params(axis='y', labelsize=12)


plt.show()
```

### Pertanyaan 2: Daerah mana yang memiliki tingkat polusi paling baik selama 2013 - 2017 bila dilihat dari kadar PM2.5?

 Visualisasi menunjukkan rata-rata kadar PM2.5 di station Dingling memiliki nilai yang paling kecil dibandingkan station lainnya

```python
fig, ax = plt.subplots(figsize=(15, 6))

colors = ['#D3D3D3', '#72BCD4', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3']

sns.barplot(x='PM2.5', y='station', data=average_all_pm25_df, palette=colors)
ax.set_ylabel('Station')
ax.set_xlabel('PM2.5', fontsize=15)
ax.set_title('Average PM2.5 in All Station', loc='center', fontsize=20)
ax.tick_params(axis='y', labelsize=12)

plt.show()
```

**Insight:**
- Visualisasi berupa horizontal bar chart untuk memvisualisasikan station mana yang memiliki kualitas udara terbaik
- Visualisasi berupa vertikal bar chart untuk memvisualisasikan perbandingan kualitas udara pada tahun 2015 dan 2016 di station Dongsi

## Conclusion

- 2016 memiliki kualitas udara yang lebih baik dari 2015 di Dongsi dengan kadar rata-rata PM2.5 sebesar 80.225952
- Daerah dengan kualitas udara terbaik selama periode 2013 - 2017 berdasarkan hasil rata-rata kadar PM2.5 adalah Dingling dengan kadar PM2.5 sebesar 65.989497
